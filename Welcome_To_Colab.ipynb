```
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5])
y=np.array([10,12,15,18,20])

plt.plot(x,y,marker='o',color='blue',linestyle='--')

plt.title("simple line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```
import seaborn as sns
import pandas as pd

data=pd.DataFrame({
    'fruits':['Apple','Banana','Cherry','Date','Avacado','Kiwi','Plums'],
    'quantity':[7,12,13,14,18,4,20]
})

sns.barplot(x='fruits',y='quantity',data=data,palette='viridis')

plt.title("Fruits in the Box")
plt.show()
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df=sns.load_dataset("iris")

plt.figure(figsize=(6,4))
sns.histplot(data=df,x="sepal_length",bins=20,kde=True,color='blue')
plt.title("Histogram of sepal length by species")
plt.show()
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df=sns.load_dataset("iris")

plt.figure(figsize=(6,4))
sns.boxplot(data=df,x="species",y="sepal_length",palette="muted")
plt.title("Boxplot of sepal length by species")
plt.show()
```
